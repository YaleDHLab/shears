```python
import matplotlib.pyplot as plt
import glob, sys
[sys.path.append(i) for i in ['.', '..', '../picsnip']]
import picsnip

# identify the images to process
input_images = glob.glob('../sample-data/life/*.png')

# process each of the images
for i in input_images:

  # show the original page scan
  picsnip.plot_image(i, 'Page Scan')
  
  # because the image is essentially binary, do not remove colors
  # to remove the text, set a filter_min_size value
  result = picsnip.clip(i,
                        filter_min_size=2400,
                        n_colors_to_remove=0)

  # show the extracted image
  picsnip.plot_image(result, 'Extracted Image')
```
